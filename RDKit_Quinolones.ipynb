# Quinolones

Quinolones은 미생물의 DNA gyrase와 topoisomerase4를 억제하여 DNA 합성을 저해하는 항균제이다. Topoisomerase2는 인간에 존재하지만, quinolones과 결합하지 않아서 quinolones는 세균에 대한 선택적인 독성을 갖는다.

이 약물은 금속이온과 킬레이트를 형성하여 용해도가 떨어지고 약물흡수가 감소하기 때문에 유제품 또는 제산제와 병용투여는 금기이다.

처음 개발된 quinolones은 주로 그람 음성균 일부에만 항균력이 있었으나 점차 항균범위가 넓은 약물이 소개되었고, 최근에는 그람양성균과 혐기성균 등에도 항균력이 있는 quinolones이 개발되었다 

Structure of bacterial DNA gyrase complexed with DNA and two ciprofloxacin molecules (green)
- Quinolones are chemotherapeutic bactericidal drugs. They interfere with DNA replication by preventing bacterial DNA from unwinding and duplicating. Specifically, they inhibit the ligase activity of the type II topoisomerases, DNA gyrase and topoisomerase IV, which cut DNA to introduce supercoiling, while leaving nuclease activity unaffected. With the ligase activity disrupted, these enzymes release DNA with single- and double-strand breaks that lead to cell death.

```python
Quinolones=Chem.MolFromSmiles('**1c(*)c(*)c(*)c2c(=O)c(C(=O)O)c(*)n(*)c12 |$R8;X;;R7;;R6;;R5;;;;;;;;;R7;;R1$|')
Quinolones
```

1. 약효에 필수적인 pharmacophore는 carboxy-4-pyridone 구조이고, carboxylic acid와 ketone 부위가 효소와 결합하여 효소를 억제한다.
2. R6위치에 fluroro기를 도입하면, 지용성이 증가하여 세균 세포벽 투과성이 증가하고 항균력이 향상된다.
3. 1번 위치에 alkyl기를 도입하면, 약효가 증가하지만, 2번 위치에 치환체를 도입하면 약효가 감소한다.
4. R7위치에 헤테로기를 도입하면, 그람음성균에대한 항균력이 증가하나, piperazinyl기를 도입하면, CNS의 GABA receptor에 결합하여 CNS에 부작용이 발생한다.

```python
Ciprofloxacin=Chem.MolFromSmiles('C1CC1N2C=C(C(=O)C3=CC(=C(C=C32)N4CCNCC4)F)C(=O)O')
Norfloxacin=Chem.MolFromSmiles('CCN1C=C(C(=O)C2=CC(=C(C=C21)N3CCNCC3)F)C(=O)O')
Ofloxacin=Chem.MolFromSmiles('C[CH]1COC2=C3N1C=C(C(=O)C3=CC(=C2N4CCN(CC4)C)F)C(=O)O')
Levofloxacin=Chem.MolFromSmiles('C[C@H]1COC2=C3N1C=C(C(=O)C3=CC(=C2N4CCN(CC4)C)F)C(=O)O')
Moxifloxacin=Chem.MolFromSmiles('COC1=C2C(=CC(=C1N3C[C@@H]4CCCN[C@@H]4C3)F)C(=O)C(=CN2C5CC5)C(=O)O')

mols = [Quinolones, Norfloxacin,Ciprofloxacin,Ofloxacin,Levofloxacin,Moxifloxacin ]
mols_str= ["Quinolones", "Norfloxacin","Ciprofloxacin","Ofloxacin","Levofloxacin","Moxifloxacin"]
rdkit.Chem.Draw.MolsToGridImage(mols, molsPerRow=3, subImgSize=(300, 300), legends=mols_str)
```

### 약효에 필수적인 pharmacophore는 carboxy-4-pyridone 구조이고, carboxylic acid와 ketone 부위가 효소와 결합하여 효소를 억제한다.

```python
Ciprofloxacin.__sssAtoms = [0,1,2]
Ciprofloxacin
```

# Statins

HMG-CoA reductase 저해제는 HMG-CoA에서 mevalonic acid로 전환되는 것을 효과적으로 저해함으로써 혈장 콜레스테롤 수치를 낮춰준다.

```python
#Mevastatin
Mevastatin = Chem.MolFromSmiles("CC[C@H](C)C(=O)O[C@H]1CCC=C2[C@H]1[C@H]([C@H](C=C2)C)CC[C@@H]3C[C@H](CC(=O)O3)O")

#Lovastatin
Lovastatin = Chem.MolFromSmiles("CC[C@H](C)C(=O)O[C@H]1C[C@H](C=C2[C@H]1[C@H]([C@H](C=C2)C)CC[C@@H]3C[C@H](CC(=O)O3)O)C")

#Simvastatin
Simvastatin = Chem.MolFromSmiles("CCC(C)(C)C(=O)O[C@H]1C[C@H](C=C2[C@H]1[C@H]([C@H](C=C2)C)CC[C@@H]3C[C@H](CC(=O)O3)O)C")

#Pravastatin
Pravastatin = Chem.MolFromSmiles("CC[C@H](C)C(=O)O[C@H]1C[C@@H](C=C2[C@H]1[C@H]([C@H](C=C2)C)CC[C@H](C[C@H](CC(=O)O)O)O)O")
```

## HMG-COA Reductase 저해제의 구조-활성 상관관계

- 3,5-dihydroxyheptanoate는 중요한 약리단으로서 화합물의 lactone은 가수분해되어 약효를 나타내는 전구약물(prodrug)이다.
- 3번과 5번의 알코올 탄소의 절대구조는 약효에 중요하여 mevastatin과 lovastatin의 절대구조인 (3R,5R)과 동일해야한다.
- 5번 위치의 알코올 탄소와 ring system사이는 탄소 2개가 가장 이상적이다.
- Ring type 1에서는 decaline system이 가장 이상적이고 cyclohexane으로 치환할 경우 약효는 10,000배 감소한다, ester side chain의 절대구조는 크게 중요하지 않으나 eher로 변형 시 약효가 감소한다. 또한 R2가 메틸기로 치환되는 경우에는 효능이 상승한다.
- Ring type 2에서는 5환 구조나 6환 구조나 약효에서 큰 차이가 없고 X,Y,Z는 탄소나 질소이어야 한다.

## 합성된 HMG-CoA reductase 저해제

```python
#Atorvastatin
Atorvastatin = Chem.MolFromSmiles("CC(C)C1=C(C(=C(N1CC[C@H](C[C@H](CC(=O)O)O)O)C2=CC=C(C=C2)F)C3=CC=CC=C3)C(=O)NC4=CC=CC=C4")

#Fluvastatin
Fluvastatin = Chem.MolFromSmiles("CC(C)N1C2=CC=CC=C2C(=C1/C=C/[C@H](C[C@H](CC(=O)O)O)O)C3=CC=C(C=C3)F")

#Cerivastatin
Cerivastatin = Chem.MolFromSmiles("CC(C)C1=C(C(=C(C(=N1)C(C)C)COC)C2=CC=C(C=C2)F)/C=C/[C@H](C[C@H](CC(=O)O)O)O")

#Rosuvastatin calcium
Rosuvastatin_calcium = Chem.MolFromSmiles("CC(C1=NC(=NC(=C1/C=C/[C@@H](O)C[C@@H](O)CC(=O)[O-])C2=CC=C(C=C2)F)N(S(=O)(=O)C)C)C.CC(C1=NC(=NC(=C1/C=C/[C@@H](O)C[C@@H](O)CC(=O)[O-])C2=CC=C(C=C2)F)N(S(=O)(=O)C)C)C.[Ca+2]")
```

# Phenothiazine

- typical antipsychotics에 포함
- 뇌의 dopamine 수용체(특히 D2)길항
- 추체외로 증후군(EPS) 부작용

N-10과 amine 사이에는 3개의 carbon사슬이 활성에 요구된다(2개의 carbon사슬이 결합된 경우는 항히스타민 작용이 있다.).
- 1. alkylaminopropyl기가 붙으면 항정신병, 진정, 항히스타민, 항콜린 작용을 나타낸다.
- 2. alkylpiperidine기가 붙으면 항 구토작용이 약하다.
- 3. alkylpiperazine은 가장 강력한 항정신병과 항구토작용이 있다.

Phenothiazine에 치환체의 도입은 EWG치환체가 2번 위치에 도입되었을 때 효능이 가장 증가한다. 
2번 위치의 중요성은 phenothiazine 골격의 곁사슬 질소에 수소가 도입되어 2번 위치의 치환체 원소의 전자쌍과 수소결합으로 연결되어 도파민과 유사한 배향을 이루기 때문이다.

N-10과 amine 사이 3개의 carbon사슬이 활성에 중요하다.
- 이 탄소에 더 큰 alkyl기를 붙이면 활성이 줄어든다.
- 또한 탄소원자가 자유롭게 회전할 수 있는 것이 좋다.

```python
Chlorpromazine = Chem.MolFromSmiles("CN(C)CCCN1C2=CC=CC=C2SC3=C1C=C(C=C3)Cl")
Chlorpromazine
```

```python
Fluphenazine = Chem.MolFromSmiles("C1CN(CCN1CCCN2C3=CC=CC=C3SC4=C2C=C(C=C4)C(F)(F)F)CCO")
Fluphenazine
```

```python
Prochloperazine = Chem.MolFromSmiles("CN1CCN(CC1)CCCN2C3=CC=CC=C3SC4=C2C=C(C=C4)Cl")
Prochloperazine
```

```python
Dopamine = Chem.MolFromSmiles("C1=CC(=C(C=C1CCN)O)O")
Dopamine
```

```python
from rdkit.Chem.Draw import SimilarityMaps
fig, maxweight = SimilarityMaps.GetSimilarityMapForFingerprint(Dopamine, Chlorpromazine, SimilarityMaps.GetMorganFingerprint)
```

SimilarityMaps를 통해 dopamine과 chlorpromazine의 구조적 유사성을 찾기 힘드나
phenothiazine 골격의 곁사슬 질소에 결합된 수소와 2번위치의 치환체 원소의 전자쌍과 수소결합으로 연결되어 도파민과 유사한 배향을 이룬다.

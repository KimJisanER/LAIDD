# Benzodiazepines

Benzodiazepines are depressants that enhance the effect of the neurotransmitter gamma-aminobutyric acid (GABA) at the GABAA receptor, resulting in sedative, hypnotic (sleep-inducing), anxiolytic (anti-anxiety), anticonvulsant, and muscle relaxant properties.

< A ring >
- 방향족 필요(hetero : 활성 감소), 수용체와 π-π 상호작용
- 7번 C에만 EWG 치환 시 활성

< B ring >
- C2 : proton acceptor 필요( His와결합 )
- C1-C2 : triazole/imidazole이면 활성 증가
- C3 : -OH 도입 시 대사 빨라짐(lorazepam)
- N : 큰 치환기 존재 시 친화성 감소
- N4-C5 : 포화 시 활성 증가

< C ring >
- 필수보단 최적화에 유리
- C2: EWG 존재 시 활성 증가
- C4: 치환 시 활성 감소

```python
import rdkit
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem.Draw import IPythonConsole
IPythonConsole.drawOptions.addAtomIndices = False
IPythonConsole.drawOptions.addStereoAnnotation = True
IPythonConsole.molSize = 350,350
```

* Maximum Common Substructure (MCS) 는 여러개의 분자가 모두 가지고 있는 common 구조를 얻어내는 것을 의미한다. 
* Chem 모듈 안의 rdFMCS 모듈을 import

```python
from rdkit.Chem import rdFMCS
```

# Types
* 2-keto compounds:
clorazepate, diazepam, flurazepam
* 3-hydroxy compounds:
lorazepam
* 7-nitro compounds:
clonazepam
* Triazolo compounds:
alprazolam, estazolam, triazolam
* Imidazo compounds:
midazolam
* 1,5-benzodiazepines:
clobazam

```python
#Clorazepate
Clorazepate = Chem.MolFromSmiles("C1=CC=C(C=C1)C2=NC(C(=O)NC3=C2C=C(C=C3)Cl)C(=O)O")
#Diazepam
Diazepam = Chem.MolFromSmiles("CN1C(=O)CN=C(C2=C1C=CC(=C2)Cl)C3=CC=CC=C3")
#Flurazepam
Flurazepam = Chem.MolFromSmiles("CCN(CC)CCN1C(=O)CN=C(C2=C1C=CC(=C2)Cl)C3=CC=CC=C3F")

#Lorazepam
Lorazepam = Chem.MolFromSmiles("C1=CC=C(C(=C1)C2=NC(C(=O)NC3=C2C=C(C=C3)Cl)O)Cl")

#Clonazepam
Clonazepam = Chem.MolFromSmiles("C1C(=O)NC2=C(C=C(C=C2)[N+](=O)[O-])C(=N1)C3=CC=CC=C3Cl")

#Alprazolam
Alprazolam = Chem.MolFromSmiles("CC1=NN=C2N1C3=C(C=C(C=C3)Cl)C(=NC2)C4=CC=CC=C4")

#Estazolam
Estazolam = Chem.MolFromSmiles("C1C2=NN=CN2C3=C(C=C(C=C3)Cl)C(=N1)C4=CC=CC=C4")
#Triazolam
Triazolam = Chem.MolFromSmiles("CC1=NN=C2N1C3=C(C=C(C=C3)Cl)C(=NC2)C4=CC=CC=C4Cl")

#Midazolam
Midazolam = Chem.MolFromSmiles("CC1=NC=C2N1C3=C(C=C(C=C3)Cl)C(=NC2)C4=CC=CC=C4F")

#Clobazepam
Clobazepam = Chem.MolFromSmiles("CN1C(=O)CC(=O)N(C2=C1C=CC(=C2)Cl)C3=CC=CC=C3")
```

```python
mols = [Clorazepate, Diazepam, Flurazepam, Lorazepam, Clonazepam, Alprazolam, Estazolam, Triazolam, Midazolam, Clobazepam]
mols_str= ["Clorazepate", "Diazepam", "Flurazepam", "Lorazepam", "Clonazepam", "Alprazolam", "Estazolam", "Triazolam", "Midazolam", "Clobazepam"]
rdkit.Chem.Draw.MolsToGridImage(mols, molsPerRow=3, subImgSize=(200, 200), legends=mols_str)
```

```python
from rdkit.Chem.Crippen import MolLogP
for m in mols:
    logp = MolLogP(m)
    print(logp)
    m.SetProp("LogP", f"logP: {logp:.3f}") # property 설정할 때는 string으로 넣어주어야 한다.
```

```python
img=Draw.MolsToGridImage(mols,
                         molsPerRow=4,
                         subImgSize=(300, 300),
                         legends=[x.GetProp("LogP") for x in mols])
```

```python
img
```

```python
patt=Chem.MolFromSmiles("C1=NN=C2N1C3=C(C=C(C=C3)Cl)C(=NC2)C4=CC=CC=C4")
```

```python
matches = []
for mol in mols:
    if mol.HasSubstructMatch(patt):
        matches.append(mol)
        #print(Chem.MolToSmiles(mol))
    else:
        print("False:", Chem.MolToSmiles(mol))
len(matches)
```

```python
img=Draw.MolsToGridImage(matches,
                         molsPerRow=3,
                         subImgSize=(300, 300),
                         legends=["Alprazolam", "Estazolam", "Triazolam"])
```
